```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision

import random
import math

#import pandas
```

```python
def simple_training(data, learning_rate, batch_size, epochs):
    model = torch.nn.Sequential(
        torch.nn.Linear(28*28, 256),
        torch.nn.ReLU(),
        torch.nn.Linear(256, 10)
    )
    
    opt = torch.optim.Adam(params=model.parameters(), lr=learning_rate)

    loss_fn = torch.nn.CrossEntropyLoss()
    
    train_loader = torch.utils.data.DataLoader(data, batch_size=batch_size, shuffle=True)

    losses = []
    for epoch in range(epochs):
        for sample, labels in train_loader:
            n = len(sample)
            # Reshape data from [500, 1, 28, 28] to [500, 784] and use the model to make predictions.
            predictions = model(sample.view(n, -1))
            # Compute the loss.
            loss = loss_fn(predictions, labels)
            opt.zero_grad()
            loss.backward()
            opt.step()
            losses.append(float(loss))
        print(f"Epoch: {epoch}, Loss: {float(loss)}")
            
```
